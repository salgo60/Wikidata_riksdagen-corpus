## Get data from Wikidata and check quality of Riksdagens data

```python

import sys
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

# https://w.wiki/6hRT
query = """SELECT (REPLACE(STR(?person), ".*Q", "Q") AS ?qid) ?personLabel  ?wd_iort ?Riksdagenguid ?RiksdagenOldid {
  values ?befattning  {wd:Q110346241 wd:Q81531912 wd:Q10655178} # 1st 2nd todays
  
 
  ?person wdt:P39 ?befattning.
  
  OPTIONAL {?person wdt:P8433 ?Riksdagenguid} 
  OPTIONAL {?person wdt:P1214 ?RiksdagenOldid}
  
  OPTIONAL {?person p:P2561 ?nodeOIT.
  ?nodeOIT ps:P2561 ?wd_iort.
  ?nodeOIT pq:P3831 wd:Q110382440  # I riksdagen kallad 
                       }
    SERVICE wikibase:label { bd:serviceParam wikibase:language "sv,en". }
} order by ?personLabel"""

    
def get_sparql_dataframe(endpoint_url, query):
    """
    Helper function to convert SPARQL results into a Pandas data frame.
    """
    user_agent = "salgo60/%s.%s" % (sys.version_info[0], sys.version_info[1])
 
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    result = sparql.query()

    processed_results = json.load(result.response)
    cols = processed_results['head']['vars']
    out = []
    for row in processed_results['results']['bindings']:
        item = []
        for c in cols:
            item.append(row.get(c, {}).get('value'))
        out.append(item)

    return pd.DataFrame(out, columns=cols)

WDRiksdagen = get_sparql_dataframe(endpoint_url, query)

WDRiksdagen.info()
```

```python
#Riksdageniort_WD_merge =  pd.merge(dfiort, WDRiksdagen, left_on='intressent_id',right_on='RiksdagenOldid',
#                                  indicator=True, how="outer")
Riksdageniort_WD_merge =  pd.merge(dfiort, WDRiksdagen, left_on='intressent_id',right_on='RiksdagenOldid',
                                   indicator=True)

Riksdageniort_WD_merge.info()
```
