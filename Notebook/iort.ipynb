### Get iort from Riksdagens öppna data
* see [#141](https://github.com/salgo60/Wikidata_riksdagen-corpus/issues/141) 

```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
url ="http://data.riksdagen.se/personlista/?rdlstatus=samtliga&utformat=json&termlista="

r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data["personlista"]["person"])
dftot.head()
```

```python
dftot.info()
```

```python
dftot.iort.value_counts()
```

```python
dfiort = dftot[dftot.iort.str.len() > 0]
```

```python
dfiort.iort
```

```python
dfiort
```

```python
dfiort[["iort","status","sorteringsnamn","sourceid","intressent_id"]]
```

```python
dfiort[["iort","status","sorteringsnamn","sourceid","intressent_id"]].to_csv("iort2023maj.csv")
```
